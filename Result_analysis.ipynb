```python
import pandas as pd

# File path
file_path = './Data/Result.csv'

# Read the file as a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first few rows
#print(df.sample(2))
```

```python
selected_columns = ['3', '5', '10', '15', '20', '25', '30', '35', '40', '45']
# Convert selected columns to numeric (if not already)
df[selected_columns] = df[selected_columns].apply(pd.to_numeric, errors='coerce')

# Calculate statistics
df['Maximum'] = df[selected_columns].max(axis=1)
df['Average'] = df[selected_columns].mean(axis=1)
df['Median'] = df[selected_columns].median(axis=1)
df['Minimum'] = df[selected_columns].min(axis=1)
df['Standard Deviation'] = df[selected_columns].std(axis=1)

# Save or display the updated DataFrame
#print(df.head())

# Display the updated DataFrame in the console
# Exclude the selected columns and display the result
result = df.drop(columns=selected_columns)
print(result)

# Save the updated DataFrame to a new CSV file
output_path = './Data/all_dataset_result.csv'
result.to_csv(output_path, index=False)
print(f"Updated file saved to {output_path}")
```

```python
# Filter the dataset for "YT-Title" and "20-Newsgroup"
filtered_df = df[df['Dataset'].isin(['YT-Title', '20-Newsgroup'])]

# Define columns to drop (selected + additional)
selected_columns = ['5', '10', '20', '30', '40']
additional_columns_to_drop = ['3', '15', '25', '35', '45']
columns_to_drop = selected_columns + additional_columns_to_drop

# Convert selected columns to numeric (if not already)
filtered_df[selected_columns] = filtered_df[selected_columns].apply(pd.to_numeric, errors='coerce')

# Calculate statistics
filtered_df['Maximum'] = filtered_df[selected_columns].max(axis=1)
filtered_df['Average'] = filtered_df[selected_columns].mean(axis=1)
filtered_df['Median'] = filtered_df[selected_columns].median(axis=1)
filtered_df['Minimum'] = filtered_df[selected_columns].min(axis=1)
filtered_df['Standard Deviation'] = filtered_df[selected_columns].std(axis=1)

# Drop the specified columns from the filtered dataset
result = filtered_df.drop(columns=columns_to_drop)

# Display the final result
print(result)

# Save the updated DataFrame to a new CSV file
output_path = './Data/large_dataset_result.csv'
result.to_csv(output_path, index=False)
print(f"Updated file saved to {output_path}")
```
